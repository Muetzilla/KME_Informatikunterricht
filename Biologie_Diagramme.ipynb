# 2D Zwischenartliche Konkurrenz

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Baumarten mit Standortdaten
arten = ['Birke', 'Hainbuche', 'Lärche']
werte = {
    'Birke': {
        'existenz': [10, 20, 90, 90],       # [x_min, y_min, x_max, y_max]
        'bevorzugt': [20, 30, 70, 70],
        'vorkommen': [25, 40, 60, 60]
    },
    'Hainbuche': {
        'existenz': [30, 40, 90, 80],
        'bevorzugt': [40, 50, 80, 70],
        'vorkommen': [50, 50, 85, 70]
    },
    'Lärche': {
        'existenz': [0, 10, 60, 60],
        'bevorzugt': [10, 20, 50, 50],
        'vorkommen': [15, 20, 40, 40]
    }
}

# Schraffuren und Farben für jede Art
hatch_patterns = {
    'Birke': '///',
    'Hainbuche': '\\\\\\',
    'Lärche': 'xxx'
}
edge_colors = {
    'Birke': 'red',
    'Hainbuche': 'purple',
    'Lärche': 'brown'
}

# Diagramm vorbereiten
fig, ax = plt.subplots(figsize=(10, 8))
ax.set_xlim(0, 100)
ax.set_ylim(0, 100)
ax.invert_yaxis()  # Feucht unten, trocken oben
ax.set_xlabel("Bodensäuregrad (← basisch | sauer →)", fontsize=12)
ax.set_ylabel("Bodenfeuchtigkeit (trocken ↑ | nass ↓)", fontsize=12)
ax.set_title("Vergleich der Standortansprüche und tatsächlichen Vorkommen", fontsize=14)

# Legendenliste
handles = []

# Daten einzeichnen
for art in arten:
    daten = werte[art]

    # Existenzbereich (grün transparent)
    existenz = patches.Rectangle(
        (daten['existenz'][0], daten['existenz'][1]),
        daten['existenz'][2] - daten['existenz'][0],
        daten['existenz'][3] - daten['existenz'][1],
        linewidth=1,
        edgecolor='green',
        facecolor='green',
        alpha=0.1
    )
    ax.add_patch(existenz)

    # Bevorzugter Bereich (orange)
    bevorzugt = patches.Rectangle(
        (daten['bevorzugt'][0], daten['bevorzugt'][1]),
        daten['bevorzugt'][2] - daten['bevorzugt'][0],
        daten['bevorzugt'][3] - daten['bevorzugt'][1],
        linewidth=1,
        edgecolor='orange',
        facecolor='orange',
        alpha=0.3
    )
    ax.add_patch(bevorzugt)

    # Tatsächliches Vorkommen (schraffiert und farbig)
    vorkommen = patches.Rectangle(
        (daten['vorkommen'][0], daten['vorkommen'][1]),
        daten['vorkommen'][2] - daten['vorkommen'][0],
        daten['vorkommen'][3] - daten['vorkommen'][1],
        linewidth=2,
        edgecolor=edge_colors[art],
        facecolor='none',
        hatch=hatch_patterns[art],
        label=f"{art} (Vorkommen)"
    )
    ax.add_patch(vorkommen)
    handles.append(vorkommen)

# Legende einfügen
ax.legend(handles=handles, loc='lower left', fontsize=10)

plt.tight_layout()
plt.show()
```

# Zwischenartliche Konkurrenz 2D Einzelne Bäume

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Funktion zum Zeichnen je Art
def zeichne_baumdiagramm(ax, artname, existenz, bevorzugt, vorkommen=None):
    ax.set_xlim(0, 100)
    ax.set_ylim(0, 100)
    ax.set_xlabel("Bodensäuregrad (← basisch | sauer →)")
    ax.set_ylabel("Bodenfeuchtigkeit (trocken ↑ | nass ↓)")
    ax.set_title(artname, fontsize=12)
    ax.invert_yaxis()

    # Existenzbereich (grün)
    ax.add_patch(patches.Rectangle((existenz[0], existenz[1]),
        existenz[2] - existenz[0], existenz[3] - existenz[1],
        edgecolor='green', facecolor='green', alpha=0.2))

    # Bevorzugter Bereich (orange)
    ax.add_patch(patches.Rectangle((bevorzugt[0], bevorzugt[1]),
        bevorzugt[2] - bevorzugt[0], bevorzugt[3] - bevorzugt[1],
        edgecolor='orange', facecolor='orange', alpha=0.6))

    # Tatsächliches Vorkommen (rot schraffiert)
    if vorkommen:
        ax.add_patch(patches.Rectangle((vorkommen[0], vorkommen[1]),
            vorkommen[2] - vorkommen[0], vorkommen[3] - vorkommen[1],
            edgecolor='red', facecolor='none', hatch='///', linewidth=2))

# Daten für 3 Baumarten
arten = ['Birke', 'Hainbuche', 'Lärche']
werte = {
    'Birke': {'existenz': [10, 20, 90, 90], 'bevorzugt': [20, 30, 70, 70], 'vorkommen': [25, 40, 60, 60]},
    'Hainbuche': {'existenz': [30, 40, 90, 80], 'bevorzugt': [40, 50, 80, 70], 'vorkommen': [50, 50, 85, 70]},
    'Lärche': {'existenz': [0, 10, 60, 60], 'bevorzugt': [10, 20, 50, 50], 'vorkommen': [15, 20, 40, 40]}
}

# Zeichnen
fig, axs = plt.subplots(1, 3, figsize=(16, 6))
for ax, art in zip(axs, arten):
    zeichne_baumdiagramm(ax, art, werte[art]['existenz'], werte[art]['bevorzugt'], werte[art]['vorkommen'])

fig.suptitle("Ansprüche und tatsächliches Vorkommen von Baumarten im Mischwald", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

# Zwischenartliche Konkurrenz

```python
import matplotlib.pyplot as plt
import numpy as np

# Hilfsfunktion für Toleranzkurven
def gaussian(x, mu, sigma):
    return np.exp(-0.5 * ((x - mu) / sigma)**2)

# Umweltfaktor: Feuchtegradient von 0 (trocken) bis 100 (sehr feucht)
feuchte = np.linspace(0, 100, 300)

# Reinsituation (ohne Konkurrenz)
heidelerche_rein = gaussian(feuchte, 30, 25)
buchfink_rein = gaussian(feuchte, 50, 25)
teichrohrsaenger_rein = gaussian(feuchte, 70, 25)

# Mischsituation (mit Konkurrenz)
heidelerche_misch = gaussian(feuchte, 20, 10)
buchfink_misch = gaussian(feuchte, 50, 15)
teichrohrsaenger_misch = gaussian(feuchte, 85, 10)

# Plot 1 – Reinsaat
plt.figure(figsize=(10, 5))
plt.plot(feuchte, heidelerche_rein, label='Heidelerche', color='green')
plt.plot(feuchte, buchfink_rein, label='Buchfink', color='orange')
plt.plot(feuchte, teichrohrsaenger_rein, label='Teichrohrsänger', color='blue')
plt.title('Reinsituation (ohne Konkurrenz)', fontsize=14)
plt.xlabel('Feuchte des Lebensraums')
plt.ylabel('Häufigkeit / Eignung')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot 2 – Mischsituation
plt.figure(figsize=(10, 5))
plt.plot(feuchte, heidelerche_misch, label='Heidelerche', color='green')
plt.plot(feuchte, buchfink_misch, label='Buchfink', color='orange')
plt.plot(feuchte, teichrohrsaenger_misch, label='Teichrohrsänger', color='blue')
plt.title('Mischsituation (mit Konkurrenz)', fontsize=14)
plt.xlabel('Feuchte des Lebensraums')
plt.ylabel('Häufigkeit / Durchsetzung')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
